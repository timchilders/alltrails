```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
#plotting
color_list = plt.cm.summer(np.linspace(1, 0, 12))
```

Load the two csv files of Alltrails data taken from https://github.com/oschow/take-a-hike/tree/master/AllTrails/data

```python
trail_df = pd.read_csv('../data/all_hikes_with_hike_name.csv')
ratings_df = pd.read_csv('../data/all_ratings_matrix.csv')
```

Let's look at what data is in the all_hikes_with_hike_name.csv.

```python
trail_df.head()
```

Are there any null values in the data?

```python
print(trail_df.isnull().sum(axis=0),ratings_df.isnull().sum(axis=0))
```

The data looks clean but it is missing some valuable information that could have been scraped, including: latitude and longitude, additional keywords, the number of photos taken, the number of reviews and the average rating. Fortunately all_ratings_matrix.csv contains ratings for each trail. Let's combine these two df's into one.

```python
ratings_groups = ratings_df.groupby('hike_id')
nratings = ratings_groups.count()
nratings = nratings.drop('value',axis=1)
meanrating = ratings_groups.mean()
ratings_by_hike = pd.concat([nratings,meanrating],axis=1)
ratings_by_hike = ratings_by_hike.rename(columns={'variable':'reviews','value':'mean rating'})
```

```python
df = trail_df.merge(ratings_by_hike,on='hike_id')
df
```

We now have one table with data for 1487 trails in Colorado, including the number of reviews per trail and their mean value.
Let's quickly look at other stats for the dataset.

```python
df.describe()
```

```python
df.hist(figsize=(12,12))
```

What are the most popular trails in Colorado (most reviews)?

```python
# reviews = df[['hike_name','reviews']].sort_values(by='reviews',ascending=False).set_index(df['hike_name'])
# reviews.iloc[:15].plot(kind='bar')
```

What are the most popular trail features?

```python
tags = df.loc[:,'loop':'views']
tag_tot = tags.sum().sort_values(ascending=True)
features = tag_tot.index[:]

fig,ax = plt.subplots()
ax.barh(features,tag_tot, color=color_list)
ax.set_title('Trail Features')
ax.set_xlabel('# of trails')
fig.tight_layout()

# fig.savefig('../images/trail_features.png', dpi=200)
# tag_tot.plot(kind='bar',title='Trail Features',rot=85, color=color_list)
```

Where are these trails located?

```python
region = df.groupby('hike_region')
num_trails = region.hike_name.count().sort_values(ascending=False)
locations = num_trails.iloc[:10].sort_values()
fig,ax = plt.subplots(figsize=(12,8))
ax.barh(locations.index[:],locations, color=color_list)
ax.set_title('Trail Locations')
ax.set_xlabel('# of trails')
fig.tight_layout()

# fig.savefig('../images/trail_locations.png', dpi=200)
# num_trails.iloc[:15].plot(kind='bar')
```

Let's narrow down our focus on just the ratings.

```python
ax = df['mean rating'].plot(kind='hist',bins=50,color=color_list[9],title='Trail Ratings')
ax.set_xlabel('Rating (1-5 stars)')
fig=ax.get_figure()
fig.figsize=(10,10)
fig.tight_layout()
# fig.savefig('../images/trail_ratings.png', dpi=200)
```

There seems to be outliers for integer valued ratings. This may be due to there being many trails with few reviews, but let's make sure by excluding those.

```python
ax = df[df['reviews']>5]['mean rating'].plot(kind='hist',bins=50,color=color_list[9],title='Trail Ratings(>5 total ratings)')
ax.set_xlabel('Rating (1-5 stars)')
fig=ax.get_figure()
fig.figsize=(10,10)
fig.tight_layout()
# fig.savefig('../images/trail_ratings_filtered.png', dpi=200)
```

```python
fig, axs = plt.subplots(1,2,figsize=(10,5))
axs[0].hist(df['mean rating'],bins=50,color=color_list[9])
axs[0].set_title('Trail Ratings')
axs[0].set_ylabel('Total reviews per trail')
axs[0].set_xlabel('Rating (1-5 stars)')
axs[0].vlines(x=df['mean rating'].mean(),ymin=0,ymax=250,linestyle='dashed')
axs[1].hist(df[df['reviews']>5]['mean rating'],bins=50,color=color_list[9])
axs[1].set_title('Trail Ratings (>5 total reviews)')
axs[1].set_xlabel('Rating (1-5 stars)')
axs[1].vlines(x=df[df['reviews']>5]['mean rating'].mean(),ymin=0,ymax=50,linestyle='dashed')

fig.tight_layout()
#fig.savefig('../images/trail_ratings_compared.png', dpi=200)
```

The distrubution looks more 'normal' with a distribution roughly centered right of the mean of 4.1 stars.
Now that we know what the ratings look like, lets see if we can answer our question:\
*What makes trails popular?*\
To start, let's look at trail length and elevation gain.

What is the longest hike in this data?

```python
df.iloc[df['total_distance'].argmax()]
```

```python
fig, axs = plt.subplots(1,2,figsize=(12,6))
axs[0].scatter(df['mean rating'],df['total_distance'],alpha=0.6)
axs[0].set_ylim(0,50)
axs[1].scatter(df['mean rating'],df['elevation_gain'],alpha=0.6)
axs[1].set_ylim(0,10000)
```

There appears to be some correlation between both trail length and elevation gain. Maybe by plotting a correlation matrix, we can see what variables are most correlated with each other.

```python
df=df.drop('stars',axis=1)
df.corr()
```

```python
fig, ax = plt.subplots(figsize=(12,12)) 
ax = sn.heatmap(df.corr(),annot=True,linewidths=0.1,ax=ax, cmap="Greens")
# fig.savefig('heat_map.png',dpi=200)
```

As expected, there is some positive correlation of elevation gain and trail length with how it's rated. However, this correlation does not appear to be strong. Other interesting features that stick out are a negative correlation of kid-friendly trails and ratings and a strong positive correlation of wildlife and wildflowers with views. (I'm ignoring the stars category because it is just the rounded mean rating)

## Hypothesis Forming:

*High rated hikes are more likely to be longer and more difficult and not kid friendly* 
* Null hypothesis: hike length or total elevation gain have no effect on hike rating.

*Popular hikes (most reviews) feature waterfalls and wildflowers.* 
* Null hypothesis: popular hikes have no distinctive features.
